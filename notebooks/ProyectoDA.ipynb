```python
# FOLDER --- OIL

import os
import pandas as pd

# LOAD FOLDER OF DATA

from google.colab import drive
drive.mount('/content/drive', force_remount= True)

folder_path = "/content/drive/MyDrive/EvolucionMA/Data/PRODUCCIÓN-CYG-ANH/CRUDOANH"

# Lista de archivos válidos, sorted por año
archivos = [f for f in os.listdir(folder_path) if f.endswith((".xlsx", ".xls", ".csv"))]
archivos.sort(key=lambda x: int(x.split('-')[-1].split('.')[0]))

df_list = []

for archivo in archivos:
    ruta = os.path.join(folder_path, archivo)

    try:
        # EXTENSION TYPE AND LOAD
        if archivo.endswith(".xlsx") or archivo.endswith(".xls"):
            df_temp = pd.read_excel(ruta, skiprows=10)
        elif archivo.endswith(".csv"):
            df_temp = pd.read_csv(ruta, skiprows=10)

        # Normaliza columnas (strip y lower)
        df_temp.columns = df_temp.columns.str.strip().str.lower()

        # AD COLUMN YEAR (solo el número)
        year = int(archivo.split('-')[-1].split('.')[0])
        df_temp["año"] = year

        df_list.append(df_temp)
        print(f"✓ Cargado: {archivo}")
    except Exception as e:
        print(f"✗ Error: {archivo} - {e}")
        continue

# CREATE LIST OF NAMES
for i, nombre in enumerate(archivos):
    print(i, nombre)

# RENAME ALL COLUMNS USING CAPITAL LETTER
new_names = {
    'enero': 'ENERO',
    'febrero': 'FEBRERO',
    'marzo': 'MARZO',
    'abril': 'ABRIL',
    'mayo': 'MAYO',
    'junio': 'JUNIO',
    'julio': 'JULIO',
    'agosto': 'AGOSTO',
    'septiembre': 'SEPTIEMBRE',
    'octubre': 'OCTUBRE',
    'noviembre': 'NOVIEMBRE',
    'diciembre': 'DICIEMBRE',
    'departamento': 'DEPARTAMENTO',
    'operadora': 'OPERADORA',
    'campo': 'CAMPO',
    'contrato': 'CONTRATO',
    'municipio': 'MUNICIPIO',
    'año': 'AÑO'  # Renombra a mayús
}

for df in df_list:
    df.rename(columns=new_names, inplace=True)

# DATA NORMALIZATION
# TYPES OF DATA
for df in df_list:
    print("Año:", df["AÑO"].iloc[0])
    print(df.dtypes)
    print("-----------")

# DIMENSIONS
for df in df_list:
    print("Año:", df["AÑO"].iloc[0])
    print(df.shape)
    print("-----------")

# DELETE LAST 2 ROWS (si DF >2)
df_list = [df.iloc[:-2] for df in df_list if len(df) > 2]

# VALUES NULL
print("Valores Nulos:")
for df in df_list:
    print("Año:", df["AÑO"].iloc[0])
    print(df.isna().sum())
    print("-----------")

# WATCH PARTICULAR DF
print("Vista previa primer DF (ultimas 5 filas):")
df_list[9].tail(5)
```

```python
# ====================================================================
# LIMPIEZA MASIVA: RECORRER CADA AÑO Y SUSTITUIR LOS VALORES SUCIOS
# ====================================================================
valores_sucios= ["-",""," ","  "]
meses_cols = ['ENERO','FEBRERO','MARZO','ABRIL','MAYO','JUNIO',
                  'JULIO','AGOSTO','SEPTIEMBRE','OCTUBRE','NOVIEMBRE','DICIEMBRE']

for i, df in enumerate(df_list):
    print(f"\nProcesando año: {df['AÑO'].iloc[0]}")

    # Solo las que existan en este df
    meses_en_df = [col for col in meses_cols if col in df.columns]

    print(f"   → Columnas de meses encontradas: {len(meses_en_df)}")

    # 2. Reemplazar todos los valores sucios por NaN
    df[meses_en_df] = df[meses_en_df].replace(valores_sucios, pd.NA)

    # 3. Rellenar NaN con 0
    df[meses_en_df] = df[meses_en_df].fillna(0)

    # 4. Guardar el df limpio de vuelta en la lista
    df_list[i] = df

print("\n¡LIMPIEZA COMPLETA! Todos los años procesados.")
```

```python
# =============================================
# MULTIPLICAR POR DÍAS DEL MES
# =============================================

import calendar

# Diccionario con los días de cada mes (año bisiesto = febrero 29)
def dias_del_mes(año, mes_nombre):
    mes_num = meses_cols.index(mes_nombre) + 1
    return calendar.monthrange(año, mes_num)[1]   # devuelve el último día del mes

# Recorremos cada año
for df in df_list:
    año = df['AÑO'].iloc[0]   # todos tienen el mismo año

    print(f"Procesando año {año}...")

    # Lista de columnas de meses que existen en este df
    meses = [col for col in df.columns if col in meses_cols]

    # Multiplicamos cada columna por sus días correspondientes
    for mes in meses:
        dias = dias_del_mes(año, mes)
        df[mes] = pd.to_numeric(df[mes], errors='coerce') * dias
        df[mes] = df[mes].round(0).astype('Int64')

    print(f"   → {año} convertido a producción mensual real (barriles totales por mes)")

print("\n¡Listo! Todos los DataFrames ahora tienen producción mensual real")
```

```python
#===============================================================
# Lista donde guardaremos los DataFrames en el nuevo formato
#=============================================================
df_nuevo_formato = []

for df in df_list:
    # Convertimos a formato largo
    df_temp = pd.melt(
        df,
        id_vars=['DEPARTAMENTO', 'MUNICIPIO', 'OPERADORA', 'CONTRATO', 'CAMPO', 'AÑO'],
        value_vars=[m for m in meses_cols if m in df.columns],  # trae solo los meses que existen
        var_name='MES',
        value_name='PRODUCCION'
    )

    # ¡AQUÍ GUARDAMOS!
    df_nuevo_formato.append(df_temp)

    print(f"Convertido año {df['AÑO'].iloc[0]} → {df_temp.shape[0]} filas")
```

```python
# VALUES NULL in df_nuevo_formato
print("Nulos después:")
for df in df_nuevo_formato:
    print("Año:", df["AÑO"].iloc[0])
    print(df.isna().sum())
    print("-----------")
```

```python
#======================================
#      CREAR DATAFRAME TOTAL
#======================================


df_total = pd.concat(df_nuevo_formato, ignore_index=True)


print(f"DataFrame maestro creado: {df_total.shape[0]:,} filas × {df_total.shape[1]} columnas")
df_total.tail(5)
```

```python
# =============================================
# LIMPIEZA PERFECTA: SIN TILDES
# =============================================

import unicodedata


def limpiar_sin_borrar_ñ(texto):
    if pd.isna(texto):
        return texto

    texto = str(texto)

    # Primero: protegemos la ñ y Ñ convirtiéndolas temporalmente
    texto = texto.replace('ñ', '###NYE###').replace('Ñ', '###NYE_MAYUS###')

    # Ahora sí normalizamos y quitamos tildes con seguridad
    texto = unicodedata.normalize('NFD', texto)
    texto = ''.join(c for c in texto if unicodedata.category(c) != 'Mn')

    # Devolvemos la ñ a su lugar
    texto = texto.replace('###NYE###', 'ñ').replace('###NYE_MAYUS###', 'Ñ')

    # Limpiamos y ponemos en mayúsculas (opcional)
    return texto.strip().upper()

# === Aplicar ===
print("Aplicando limpieza perfecta...")

df_total['DEPARTAMENTO'] = df_total['DEPARTAMENTO'].apply(limpiar_sin_borrar_ñ)
df_total['MUNICIPIO']    = df_total['MUNICIPIO'].apply(limpiar_sin_borrar_ñ)
df_total['CAMPO']        = df_total['CAMPO'].apply(limpiar_sin_borrar_ñ)
df_total['OPERADORA']    = df_total['OPERADORA'].apply(limpiar_sin_borrar_ñ)

# === VERIFICACIÓN 100% REAL ===
print("\nVERIFICACIÓN FINAL (debe salir con ñ):")
print("CAÑO LIMÓN →", df_total[df_total['CAMPO'].str.contains('CAÑO', case=False, na=False)]['CAMPO'].iloc[0])


print("\nDepartamentos limpios (sin tildes, con ñ):")
print("NARIÑO     →", df_total[df_total['DEPARTAMENTO'].str.contains('NARI', na=False)]['DEPARTAMENTO'].unique()[0])

```
